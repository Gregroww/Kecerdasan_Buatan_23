```python
# Menampilkan Statistik Deskriptif Dataset Books 

import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns

df = pd.read_csv('books.csv')
df_books = pd.read_csv('books.csv')
df.describe()
```

```python
total_books = df['bookID'].nunique()
total_publishers = df['publisher'].nunique()
total_authors = df['authors'].nunique()
print("Total Buku       :", total_books)
print("Total Publisher  :", total_publishers)
print("Total Author     :", total_authors)
```

```python
# Menampilkan 10 record pertama dari dataset books
df.head(10)
```

```python
# Menampilkan jumlah record,jumlah atribut, nama atribut, jumlah artibut dengan tipe angka
df.info()
```

```python
# Menampilkan jumlah label dari semua attribute yang bernilai object (contoh : Jenis Kelamin, Agama).

object_cols = df.select_dtypes(include=['object']).columns
for col in object_cols:
    print(f"Jumlah label unik pada atribut '{col}': {df[col].nunique()}")
```

```python
# Menampilkan grafik untuk avg rating buku

plt.figure(figsize=(8,5))
plt.hist(df['average_rating'], bins=30, edgecolor='black', color='skyblue')
plt.title("Distribusi Average Rating Buku")
plt.xlabel("Average Rating")
plt.ylabel("Jumlah Buku")
plt.grid(axis='y', alpha=0.5)
plt.show()
```

```python
# Menampilkan grafik untuk jumlah buku per tahun rilis

df['publication_date'] = pd.to_datetime(df['publication_date'], errors='coerce') #konfersi kolom ke datetime
df = df.dropna(subset=['publication_date'])#buang baris yang gagal parsing
books_per_year = df['publication_date'].dt.year.value_counts().sort_index()#hitung jumlah buku per tahun rilis
# Plot line chart
plt.figure(figsize=(12,6))
plt.plot(books_per_year.index, books_per_year.values, marker='o', color='royalblue')
plt.title("Jumlah Buku per Tahun Rilis")
plt.xlabel("Tahun Rilis")
plt.ylabel("Jumlah Buku")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Menampilkan grafik untuk jumlah halaman buku

plt.figure(figsize=(8,5))
plt.hist(df['  num_pages'], bins=50, color='lightcoral', edgecolor='black')
plt.title("Distribusi Jumlah Halaman Buku")
plt.xlabel("Jumlah Halaman")
plt.ylabel("Jumlah Buku")
plt.xlim(0, 1500)
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Menampilkan 5 bahasa terbanyak
lang_counts = df['language_code'].value_counts().head(5)

plt.figure(figsize=(7,7))
plt.pie(lang_counts, labels=lang_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Top 5 Bahasa Buku dalam Dataset")
plt.show()
```

```python
# Menampilkan heatmap korelasi antar atribut numerik

numeric_df = df.select_dtypes(include=['int64', 'float64'])
corr_matrix = numeric_df.corr()#hitung korelasi
# Plot heatmap
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Heatmap Korelasi Atribut")
plt.show()
```
