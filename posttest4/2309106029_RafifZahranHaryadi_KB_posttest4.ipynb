```python
#Fix Handling Outlier

#Menggunakan IQR
Q1 = df['average_rating'].quantile(0.25)
Q3 = df['average_rating'].quantile(0.75)
IQR = df['average_rating'].apply(stats.iqr)
#Menghapus outlier
df['average_rating'] = df['average_rating'][~((df['average_rating'] < (Q1 - 1.5 * IQR)) | (df['average_rating'] > (Q3 + 1.5 * IQR)))]
sns.boxplot(y=df['average_rating'], color='lightblue')
ax.set_title('Boxplot Average Rating Setelah Penanganan Outlier')
```

```python
#Membuat Model Regresi untuk Memprediksi Average Rating Buku

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import MinMaxScaler

#fitur numerik yang relevan (minimal 2)
X = df_books[['  num_pages', 'ratings_count', 'text_reviews_count']]
y = df_books['average_rating']

#normalisasi data agar skala fitur sebanding
scaler = MinMaxScaler() #mmebuat objek scaler
X_scaled = scaler.fit_transform(X) #fit dan transformasi data fitur

#split data 80% training, 20% testing
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

#Linear Regression
lr = LinearRegression() #membuat objek model
lr.fit(X_train, y_train) #latih model
y_pred_lr = lr.predict(X_test) #prediksi data testing

#evaluasi linear regression 
mse_lr = mean_squared_error(y_test, y_pred_lr) #menghitung MSE
r2_lr = r2_score(y_test, y_pred_lr) #menghitung R²

#print hasil linear regression
print("Linear Regression")
print("Mean Squared Error:", mse_lr)
print("R² Score:", r2_lr)
```

```python
#Support Vector Regression

svr = SVR(kernel='rbf')   #membuat model SVR dengan kernel rbf
svr.fit(X_train, y_train) #melatih model SVR pada data training
y_pred_svr = svr.predict(X_test)  #memprediksi nilai pada data testing

#evaluasi performa SVR
mse_svr = mean_squared_error(y_test, y_pred_svr) #menghitung MSE
r2_svr = r2_score(y_test, y_pred_svr) #menghitung R²

#menampilkan hasil SVR
print("\nSupport Vector Regressor (SVR)")
print("Mean Squared Error:", mse_svr)
print("R² Score:", r2_svr)
```

```python
#Kesimpulan:
print("\nKesimpulan:")
if r2_lr > r2_svr:
    print("Model Linear Regression memiliki performa lebih baik dengan R² Score:", r2_lr)
else:
    print("Model Support Vector Regressor memiliki performa lebih baik dengan R² Score:", r2_svr)
    print("Jika SVR lebih baik, berarti pola data non-linear dan kernel RBF menangkapnya dengan baik.")

    print("")
    print('''          Berdasarkan hasil implementasi metode Supervised Learning (Regresi) menggunakan 
          Linear Regression dan Support Vector Regressor (SVR), dapat disimpulkan bahwa 
          kedua model mampu memprediksi nilai average_rating berdasarkan fitur numerik 
          seperti num_pages, ratings_count, dan text_reviews_count.Proses splitting 
          data telah membagi dataset menjadi 80% data training dan 20% data testing, 
          kemudian dilakukan normalisasi agar seluruh fitur berada pada skala 
          yang sama sehingga model lebih stabil.''')

# Nama:Rafif Zahran Haryadi
```

```python
#evaluasi regresi

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Evaluasi Linear Regression
mae_lr = mean_absolute_error(y_test, y_pred_lr) #menghitung MAE
mse_lr = mean_squared_error(y_test, y_pred_lr)#menghitung MSE
rmse_lr = np.sqrt(mse_lr)#menghitung RMSE
r2_lr = r2_score(y_test, y_pred_lr)#menghitung R²

print("Linear Regression")
print("MAE :", mae_lr)
print("MSE :", mse_lr)
print("RMSE:", rmse_lr)
print("R2  :", r2_lr)

# Evaluasi Support Vector Regressor
mae_svr = mean_absolute_error(y_test, y_pred_svr)#menghitung MAE
mse_svr = mean_squared_error(y_test, y_pred_svr)#menghitung MSE
rmse_svr = np.sqrt(mse_svr)#menghitung RMSE
r2_svr = r2_score(y_test, y_pred_svr)#menghitung R²

print("Support Vector Regressor")
print("MAE :", mae_svr)
print("MSE :", mse_svr)
print("RMSE:", rmse_svr)
print("R2  :", r2_svr)
```

```python
#Kesimpulan:


#MAE (Mean Absolute Error)
#Rata-rata kesalahan absolut model Linear Regression adalah 0.2254, sedangkan SVR sedikit lebih kecil yaitu 0.2210.
#→ Artinya secara rata-rata, prediksi model SVR hanya meleset sekitar 0.22 poin dari nilai average_rating sebenarnya.

#MSE (Mean Squared Error)
#Nilai MSE SVR (0.1157) sedikit lebih rendah daripada Linear Regression (0.1178).
#→ menunjukkan SVR membuat kesalahan kuadrat lebih kecil dan lebih stabil dalam memprediksi.

#RMSE (Root Mean Squared Error)
#RMSE SVR = 0.3402, Linear Regression = 0.3432.
#→ SVR unggul sedikit. Rata-rata kesalahan prediksi sekitar 0.34 poin, yang masih tergolong kecil untuk skala rating 0–5.

#R² Score (Koefisien Determinasi)
#Nilai R² Linear Regression = 0.026 dan SVR = 0.043.
#→ Kedua model belum menjelaskan variasi data dengan baik (karena nilai masih di bawah 0.1).
#Namun, SVR memiliki R² lebih tinggi, artinya sedikit lebih mampu menangkap pola non-linear antara fitur dan average_rating.


#Kesimpulan Akhir:
print('''Berdasarkan hasil evaluasi menggunakan MAE, MSE, RMSE, dan R²-Score, 
      model Support Vector Regressor (SVR) menunjukkan performa 
      sedikit lebih baik dibandingkan Linear Regression. 
      Nilai error (MAE, MSE, RMSE) yang rendah menandakan bahwa hasil 
      prediksi mendekati nilai aktual average_rating, meskipun nilai R² 
      yang kecil menunjukkan bahwa model belum mampu menjelaskan variasi 
      data secara optimal. Dengan demikian, SVR lebih direkomendasikan 
      untuk dataset ini, dan performanya dapat ditingkatkan dengan menambahkan 
      fitur tambahan yang lebih relevan.''')
```
